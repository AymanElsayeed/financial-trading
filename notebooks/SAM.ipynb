# Simple Moving Average (SMA)

```python
import pandas as pd
import bt
import matplotlib.pyplot as plt
```

```python
import talib
```

```python
stock_data = pd.read_csv("../assets/GOOG.csv", index_col='Date', parse_dates=True)
```

```python
stock_data.head()
```

# Calculate the Simple Moving Average (SMA) for 10 days

```python
stock_data['SMA_10'] = talib.SMA(stock_data['Close'], timeperiod=10)
```

# Calculate the Simple Moving Average (SMA) for 50 days

```python
stock_data['SMA_50'] = talib.SMA(stock_data['Close'], timeperiod=50)
```

# Plot the SMA

```python
plt.plot(stock_data['Close'], label='Close')
plt.plot(stock_data['SMA_10'], label='SMA 10')
plt.plot(stock_data['SMA_50'], label='SMA 50')
plt.legend()
# set figure size
plt.gcf().set_size_inches(15, 8)
plt.show()
```
