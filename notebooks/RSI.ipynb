# Relative Strength Index (RSI)

The Relative Strength Index (RSI) is a momentum oscillator that measures the speed and change of price movements. RSI oscillates between zero and 100.

Traditionally, and according to Wilder, RSI is considered overbought when above 70 and oversold when below 30.

RSI can also be used to identify the general trend. When RSI is above 50, the trend is generally up. When RSI is below 50, the trend is generally down.

The default look-back period for RSI is 14, but this can be lowered to increase sensitivity or raised to decrease sensitivity.


$$ RSI = 100 - \frac{100}{1 + RS} $$

Where:
- $RS = \frac{Average Gain}{Average Loss}$ for the specified period.


```python
import pandas as pd
import matplotlib.pyplot as plt
import talib
```

```python
stock_data = pd.read_csv("../assets/GOOG.csv", index_col='Date', parse_dates=True)
```

```python
stock_data.head()
```

# Calculate RSI with the default time period of 14

```python
stock_data['RSI'] = talib.RSI(stock_data['Close'])
```

# Calculate RSI with a time period of 21

```python
stock_data['RSI_21'] = talib.RSI(stock_data['Close'], timeperiod=21)
```

```python
# Create subplots
fig, (ax1, ax2) = plt.subplots(2)
# Plot RSI with the price
ax1.set_ylabel('Price')
ax1.plot(stock_data['Close'])
ax2.set_ylabel('RSI')
ax2.plot(stock_data['RSI'], color='orangered')

ax1.set_title('Price and RSI')
plt.show()
```
