# Exponential Moving Average (EMA)

The Exponential Moving Average (EMA) is a type of moving average that gives more weight to recent prices in an attempt to make it more responsive to new information. The EMA reduces the lag by applying more weight to recent prices relative to older prices. The weighting applied to the most recent price depends on the specified period of the moving average. The shorter the EMA's period, the more weight that will be applied to the most recent price.

$EMA_{n} = (P_{n} * \alpha) + (EMA_{n-1} * (1 - \alpha))$

Where:
- $P_{n}$ = Price at time $n$
- alpha = $2 / (n + 1)$
- $EMA_{n-1}$ = EMA at time $n-1$
- $EMA_{n}$ = EMA at time $n$
- $n$ = Number of periods
- $n-1$ = Previous period
- $2$ = Smoothing factor


```python
import pandas as pd
import bt
import matplotlib.pyplot as plt
import talib
```

```python
stock_data = pd.read_csv("../assets/GOOG.csv", index_col='Date', parse_dates=True)
```

```python
stock_data.head()
```

# EMA Calculation

```python
stock_data["EMA_10"] = talib.EMA(stock_data["Close"], timeperiod=10)
```

```python
stock_data["EMA_50"] = talib.EMA(stock_data["Close"], timeperiod=50)
```

# Plotting EMA

```python
plt.plot(stock_data["Close"], label="Close Price")
plt.plot(stock_data["EMA_10"], label="EMA 10")
plt.plot(stock_data["EMA_50"], label="EMA 50")
plt.title("Exponential Moving Average (EMA)")
plt.legend()
# set figure size
plt.gcf().set_size_inches(19, 10)
plt.show()
```
