# ADX (Average Directional Index)

ADX suggest the direction of the trend and either builsh or bearish trend.

It is a combination of two other indicators, the positive directional indicator (+DI) and the negative directional indicator (-DI). The ADX is used to determine the strength of a trend, but not the direction. In general, the +DI and -DI are used to determine the direction of the trend. The ADX is used to determine the strength of the trend. The ADX is a lagging indicator, meaning that it does not necessarily predict future direction, but rather reflects the current strength of a trend.



```python
import pandas as pd
import bt
import matplotlib.pyplot as plt
import talib
```

```python
stock_data = pd.read_csv("../assets/GOOG.csv", index_col='Date', parse_dates=True)
```

```python
stock_data.head()
```

# Calculate the ADX with the default time period of 14

```python
stock_data['ADX_14'] = talib.ADX(stock_data['High'],
                                 stock_data['Low'], 
                                 stock_data['Close'])
```

# Calculate the ADX with the time period set to 21

```python
stock_data['ADX_21'] = talib.ADX(stock_data["High"],
                                 stock_data["Low"],
                                 stock_data["Close"],
                                 timeperiod=21)
```

```python
# Calculate ADX
stock_data['ADX'] = talib.ADX(stock_data['High'], stock_data['Low'], stock_data['Close'])

# Create subplots
fig, (ax1, ax2) = plt.subplots(2)

# Plot ADX with the price
ax1.set_ylabel('Price')
ax1.plot(stock_data['Close'])
ax2.set_ylabel('ADX')
ax2.plot(stock_data['ADX'], color='red')

ax1.set_title('Price and ADX')
plt.show()
```
